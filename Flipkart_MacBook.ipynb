```python
import pandas as pd
def flipkart():
    product_name = [i.text.strip() for i in soup.select('div._4rR01T')]
    price = [''.join((i.text.strip().split('₹')[1].split(','))) for i in soup.select('div._25b18c')]
    rating = [i.find_all('div')[0].text.strip() for i in soup.select('div.gUuXy-')]
    total_rate = [''.join(i.text.strip().split(' ')[0].split(',')) for i in soup.select('span._2_R_DZ')]
    total_reviews = [i.text.split('Ratings\xa0&\xa0')[1].split(' ')[0] for i in soup.select('span._2_R_DZ')]
    specs = [i.text for i in soup.select('div.fMghEO')]
    return pd.DataFrame({'Product Name':product_name,
                         'Price':price,
                         'Rating out of 5':rating,
                         'Total Ratings':total_rate,
                         'Total Reviews':total_reviews,
                         'Device Features':specs})
```
